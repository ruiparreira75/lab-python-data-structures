Solve the exercise by implementing the steps using the Python concepts of lists, dictionaries, sets, and basic input/output operations. 

````python
#1 Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
#2 Create an empty dictionary called `inventory`.
inventory = {}
````

````python
#3  Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the
#`inventory` dictionary and assign the respective quantities as values.
inventory = {"t-shirt": [], "mug": [], "hat": [], "book": [], "keychain": []}
inventory["t-shirt"] = input("How many quantities of t-shirts? ")
inventory["mug"] = input("How many quantities of mugs? ")
inventory["hat"] = input("How many quantities of hats? ")
inventory["book"] = input("How many quantities of books? ")
inventory["keychain"] = input("How many quantities of keychains? ")
print(inventory)
````

````python
#4. Create an empty set called `customer_orders`.
customer_orders = set()
````

````python
# 5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out 
# of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.
for i in range(3):
    order = input(f"Enter three products from {inventory} {i+1}. The choices are t-shirt, mug, hat, book, keychain.").strip().lower()
    if order in products:
        customer_orders.add(order)
    else:
        print(f"{order} is not on inventory. Please choose from {products}.")
# used {i+1} to start asking for product 1, and not product 0. 
````

````python
# 6. Print the products in the `customer_orders` set.
print(customer_orders)
````

````python
# 7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.
````

````python
print(inventory)
````

````python
total_products_ordered = len(customer_orders)
print (total_products_ordered)
````

````python
total_inventory = sum(int(value) for value in inventory.values())
print(total_inventory)
````

````python
percentage_ordered = total_products_ordered / total_inventory * 100
print(percentage_ordered)
````

````python
# 8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
````

````python
print("Total Products Ordered: total_products_ordered:",total_products_ordered)
print("Percentage of Products Ordered:",int(percentage_ordered),"%")
````
